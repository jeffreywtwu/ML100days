### HW6: URL配對

請抓出 Url 中的協定方式, 網址, 與埠

```
ex: Https://localhost:4200/ --> 抓取 Https, localhost, 4200
```

**所需配對文本:**
```
ftp://file_server.com:21/account/customers.xml
https://hengxiuxu.blogspot.tw/
file://localhost:4200
https://s3cur3-server.com:9999/
```

**應配對出的結果為**
```
ftp, file_server, 21
https, hengxiuxu.blogspot.tw
file, localhost, 4200
https, s3cur3-server.com, 9999
```

````python
import re
#HW01
txt1 = ''''02-27208889
04-2220-3585
(06)-2991111
(07)799-5678'''
locNumRegex = '\\b\d{2}\\b'
phoneNumRegex = '\\b\d{3}.*'
locNumLst = re.findall(locNumRegex,txt1, re.MULTILINE) #multiline mode
phoneNumLst = re.findall(phoneNumRegex,txt1, re.MULTILINE)

for locNum, phoneNum in zip(locNumLst, phoneNumLst):
    phoneNum = phoneNum.replace('-','')
    print(f'{locNum}, {phoneNum}')
````

````python
#HW02
txt2 = '''A121040176
L186856359
Z127598010
I114537095
D279884447
L186834359
D243736345
I114537095
Q146110887
D187217314
I114537095
Q243556025
Z127598010
H250077453
Q188367037'''
IDNumRegex = '\\b(?:H|D|Q)1\d+\\b' #(?:)non-capturing group
IDNumLst = re.findall(IDNumRegex, txt2, re.MULTILINE)
for IDNum in IDNumLst:
    print(IDNum)
````

````python
#HW03
txt3 = '''foobar@gmail.com
NoOneCareMe@gmail.com
SaveTheWorld@hotmail.com
zzzGroup@yahoo.com
eagle1963@gmail.com
maythefourthwithyiu@starwars.com'''
notGmailRegex = '(.*@[^gmail].*)'
notGmailLst = re.findall(notGmailRegex, txt3, re.MULTILINE)
for notGmail in notGmailLst:
    print(notGmail)
````

````python
#HW04
txt4 = '''<h1>This is a header 1</h1>
<a>This is a hyperlink</a>
<div class='test'>This is a text block</div>
<a href="https://regexisfun.com.tw/">Learning Regular Expression</a>'''
tagRegex = '(?<=\<)\w+'
tagLst = re.findall(tagRegex, txt4, re.MULTILINE)
for tag in tagLst:
    print(tag)
````

````python
#HW05
txt5 = '''.bash_profile
workShop.ai
file_folderName_num.jpg
favicon.png
IMG_002.png
IMG_003.gif
qoo.jpg.tmp
index.html
foobar.bmp
foobar.jpg
account.html
access.lock'''
gifjpgRegex = '.*\.(?:jpg|gif)$'
gifjpgLst = re.findall(gifjpgRegex, txt5, re.MULTILINE)
for gifjpg in gifjpgLst:
    print(gifjpg)
````
